```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Fir

```python
GALLONNS_PER_TICK = 1

df = pd.read_csv('nightdata/fir/pico-flow-reed_ticklist.1729900840.csv')
df['timestamp'] = pd.to_datetime(df.timestamp)
df['gpm'] = [x*60*GALLONNS_PER_TICK for x in list(df.frequency)]
df = df[df['frequency'] > 1.2]

plt.plot(df.timestamp, df.gpm)
plt.ylabel(f'GPM (assuming {GALLONNS_PER_TICK} gallons per tick)')
plt.ylim([60,100])
plt.tight_layout()
plt.show()
```
