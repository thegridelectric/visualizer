```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Fir

```python
GALLONNS_PER_TICK = 0.075

df = pd.read_csv('nightdata/oak/primary-flow_ticklist.1729947612.csv')
df['timestamp'] = pd.to_datetime(df.timestamp)
df['gpm'] = [x*60*GALLONNS_PER_TICK for x in list(df.frequency)]
display(df)

plt.plot(df.timestamp, df.gpm)
plt.ylabel(f'GPM (assuming {GALLONNS_PER_TICK} gallons per tick)')
plt.tight_layout()
plt.show()
```

# Oak

```python
GALLONNS_PER_TICK = 1

df = pd.read_csv('nightdata/oak/primary-flow_ticklist.1729911093.csv')
df['timestamp'] = pd.to_datetime(df.timestamp)
df['gpm'] = [x*60*GALLONNS_PER_TICK for x in list(df.frequency)]
plt.plot(df.timestamp, df.gpm)
plt.ylabel(f'GPM (assuming {GALLONNS_PER_TICK} gallons per tick)')
plt.ylim([60,100])
plt.tight_layout()
plt.show()
```
